##EXERCISE-1

Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E4.6

###PROGRAM CODE:

```python
def compute_pay(hour, rate):
   overtime_rate = 1.5
   if hour > 40:
      x = hour - 40
      rate_2 = rate * overtime_rate
      overtime_pay = x * rate_2
      pay = (40*rate) + overtime_pay
      return pay
   else :
      pay = hour*rate
      return pay


h = float(input("Enter Hours: "))
r = float(input("Enter Rate: "))
p = compute_pay(h, r)
print("Pay: ",p)
```

##CHALLENGE

Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

    Score           Grade
    [ 97, 100]      1.0
    [ 93,  97)      1.25
    [ 89,  93)      1.5
    [ 85,  89)      1.75
    [ 81,  85)      2.0
    [ 77,  81)      2.25
    [ 74,  77)      2.5
    [ 0,   74)      3.0

    Run the program repeatedly to test the various different values for input.
    
Adapted from [PEW] E4.6
<br>

###PROGRAM CODE:

```python
def computegrade(score):
    if (score >= 0) and (score < 74):
        return print('Grade\t3.0')
    elif (score >= 74) and (score < 77):
        return print('Grade\t2.5')
    elif (score >= 77) and (score < 81):
        return print('Grade\t2.25')
    elif (score >= 81) and (score < 85):
        return print('Grade\t2.0')
    elif (score >= 85) and (score < 89):
        return print('Grade\t1.75')
    elif (score >= 89) and (score < 93):
        return print('Grade\t1.5')
    elif (score >= 93) and (score < 97):
        return print('Grade\t1.25')
    elif (score >= 97) and (score <= 100):
        return print('Grade\t1.00')
    else :
        return print('Incorrect Input!!!')
s = int(input("Enter Score: "))
computegrade(s)
```
