##EXERCISE-1

In the code below, which will never print regardless of the value for x? Edit the code to correct the error.

```python
if x > 0 :
    print('Above 0')
elif x >  10 :
    print('Above 10')
elif x > 50 :
    print('Above 50')
elif x >  20 :
    print('Above 20')
else :
    print('Something else')
```


The code will never print 'Above 10', 'Above 50', and 'Above 20' because  their conditions (they used elif) overlaps with the first condition (x > 0) and x is always greater than 0 or less than zero, so the program will either print above 0 or something else.

###PROGRAM/EDITED CODE:

````python
x = int(input("Enter x: "))

if (x > 0) and (x < 10):
    print('Above 0 but below 10')
elif x==10:
      print('Equal to 10')
elif (x > 10) and (x < 20):
    print('Above 10 but below 20')
elif x == 20:
      print('Equal to 10')
elif (x > 20) and (x < 50):
    print('Above 20 but below 50')
elif x == 50 :
    print('Equal 50')
elif x > 50:
    print('Above to 50')
else :
    print('Something else or equa/less than to 0')
````

##EXERCISE-2

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    

###PROGRAM:

````python
hour = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))
overtime_rate = 1.5

if hour > 40:
   x = hour - 40
   rate_2 = rate * overtime_rate
   overtime_pay = x * rate_2
   pay = (40*rate) + overtime_pay
   print("Pay: ", pay)
else :
   pay = hour*rate
   print("Pay: ", pay)
````

##CHALLENGE:
