###ANSWER

1. Class vs. Object vs. Instance:
   -The class is like the blueprint of a house, the object is a specific house built according to that blueprint, then the instance is a particular house that has been constructed based on that blueprint. So the class is a blueprint or template for creating objects and an object is the specific instance of the class while the instance refers to the instantiation of a class, resulting in the creation of an object.

2. Method vs. Function:
   - A method is a function that is associated with a class and can be called using instance of the class while a function is a block of code that can be called independently.

3. Attribute vs. Variable:
   - An attribute is a characteristic or property of an object while a variable is a named storage location that holds a value.

4. The difference in focus and approach of procedural and OOP paradigms:
   - In procedural programming the focus is on writing functions or procedures which operate on data. While in object-oriented programming the focus is on the creation of objects which contain both data and functionality together.

5. Advantage of using a class over using modules which only contain a collection of data and functions:
   - Classes provide a way to encapsulate data and behavior into cohesive units, promoting better organization, abstraction, and reusability compared to modules which primarily offer a namespace for functions and data.
