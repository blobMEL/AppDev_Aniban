###ANSWER


1. Explain the advantages of writing your own modules.
- Writing your own modules promotes code organization, reusability, and encapsulation, making it easier to manage and maintain your codebase across multiple projects.

2. What is the preferred way of importing a module in Python?  
- The preferred way of importing a module in Python is using the `import` statement followed by the module name `import function_name`.

3. How do you get help on a function in a module?
- To get help on a function in a module, you can use the `help()` function.

4. Explain why and how 2 modules with the same function name can concurrently be used in the same Python script.

- Python allows two modules with the same function name to be used concurrently in the same script by leveraging namespaces, ensuring no naming conflicts arise, and allowing you to differentiate between functions using the module names as prefixes.


What are the constants in the math module?
- math.pi -  Represents the mathematical constant π (pi).
- math.e - Represents the mathematical constant e, the base of the natural logarithm.
- math.sqrt(2.0) - Represents the square root of 2.
- math.sin(math.pi/2) - Represents the Sine function of π/2 radians, which is equal to 1 in trigonometry.

What's does `math.exp` do? What about `math.pow`?
- exp() - Return e raised to the power of x.
- pow() - Return x**y (x to the power of y).

How do you calculate logarithms? Logarithms base 10? Natural Logarithms?
- Logarithms are calculated by using the base and the given number (n) to get the exponent that the base is being raised.
-Logarithms base 10 represents the exponent is used to raise the number 10 to get the given number (n).

- Natural logarithms, denoted as ln, represent the power to which the mathematical constant e (approximately 2.71828) must be raised to equal a given number.

```python
%%writefile location_module.py

def ask_location():
    """
    Asks the user for their current location.
    Returns:
        str: The user's location.
    """
    location = input("Where are you currently living? ")
    return location

def print_location(location):
    """
    Prints the user's location.
    Args:
        location (str): The user's location.
    """
    print(f"You are currently living in {location}.")
```

```python

import location_module

user_location = location_module.ask_location()
location_module.print_location(user_location)
```
